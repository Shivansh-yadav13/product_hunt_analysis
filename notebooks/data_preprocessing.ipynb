**Pricing Column**

```python
# Filling missing values of Pricing column
data.fillna({'Pricing': 'Payment Required'}, inplace=True)
```

**Note:** Products which are free, promote themselves heavily by giving free plan. Chances are products with only paid plans do not want list pricing. It could be possible that their pricing varies from customer to customer.
Considering the above, filling the missing values with `Payment Required` tag.

**Product Rating Column**

```python
# Number of null values
data['Product Rating'].isna().sum()
```

**Note:** Null values in product rating means, the prodcut has received no reviews and there is no rating provided. Hence, we will convert null values to 0.

```python
data.fillna({'Product Rating': 0}, inplace=True)
```

```python
data['Product Rating'].unique()
```

**Note:** Product ratings cannot be directly converted to float datatype, we will follow the below steps:
- Create a seperate list of product ratings from `Product Rating` column using `rating_to_int` method.
- Create a new `Product Rating Float` column.
- Assign the list to the new column.

```python
import re

def rating_to_int(rating):
    if type(rating) == str:
        pattern = re.compile("^[0-5](\.[0-9])?")
        match = pattern.search(rating)
        if match:
            return float(match.group(0))
        else:
            return 0
    return rating
```

```python
ratings = []

for r in data['Product Rating']:
    rating = rating_to_int(r)
    ratings.append(rating)
```

```python
data['Product Rating Float'] = ratings
```

**Day Rank & Week Rank Column**

```python
# Handling missing values in ranks
data['Day Rank'].isna().sum()
```

```python
data.loc[
    data['Date'] == '27-01-2024'
][['Date', 'Day Rank', 'Upvotes']]
```

```python
data.loc[
    data['Date'] == '28-01-2024'
][['Date', 'Day Rank', 'Upvotes', 'Comments', 'Reviews', 'Product Rating Float']]
```

```python
data.fillna({'Day Rank': 0}, inplace=True)
data.fillna({'Week Rank': 0}, inplace=True)
```

**Note:** On observation, we saw that, for all the products with no day and week rank, have not received any upvotes, comments or reviews. Hence, we are converting the day and week rank to 0 as well.

```python
data.loc[data['Day Rank'] == '-']
```

```python
# Converting Ranks to integer

def ranks_to_int(rank):
    if type(rank) == str:
        if '-' in rank:
            return 0
        rank = rank.replace("#", "")
        return int(rank)
    else:
        return 0

data['Day Rank'] = data['Day Rank'].apply(lambda x: ranks_to_int(x))
data['Week Rank'] = data['Week Rank'].apply(lambda x: ranks_to_int(x))
```

**Note:** We need to calculate the day of the launch using the `Date` column. This is crucial factor when analyzing the the product's launch.

The dates are in 2 different format, `YYYY-MM-DD` and `DD-MM-YYYY`, we need to first make a consistent format and then calculate the day

```python
data['Date Formatted'] = pd.to_datetime(data['Date'], format='mixed', dayfirst=True)
```

```python
data['Day'] = data['Date Formatted'].apply(lambda x: x.weekday())
```

```python
data.info()
```

**Note:** There are too many categories in Topics columns, but we can use `No. of Topics` as a feature for analysis. Some features are only given max 3 and minimum 1 column.

```python
def sum_topics(topic_1, topic_2, topic_3):
    topics = 0
    if type(topic_1) == str:
        topics += 1
    
    if type(topic_2) == str:
        topics += 1
    
    if type(topic_3) == str:
        topics += 1
    
    return topics

num_topics = []

for i, r in data.iterrows():
    num_topics.append(sum_topics(r['Topic 1'], r['Topic 2'], r['Topic 3']))
```

```python
data['Num Topics'] = num_topics
```

**Pricing Column**

Conveting pricing model to codes.

```python
data['Pricing'].value_counts()
```

```python
data['Pricing Codes'] = data['Pricing'].apply(lambda x: 0 if x == 'Free' else 1 if x == 'Free Options' else 2)
```
